Here we can observe the simulated impulse responses by visualizing them in the same way as in the Figure 1 example of the paper.

```python
%matplotlib inline 
import matplotlib.pyplot as plt
import pickle
import os
import time
import numpy as np

import measurement_tools

M = 8
K = 3
file_number = 1
```

An example of 16kHz room simulations:

```python
SAMPLING_FREQUENCY = 16000
[filters, alpha, tau] = measurement_tools.load_simulated_rirs(M, K, SAMPLING_FREQUENCY, file_number)

t = np.arange(0, filters.shape[0]/SAMPLING_FREQUENCY, 1/SAMPLING_FREQUENCY)
fig, axes = plt.subplots(M, 1, figsize=(20,10))
for i in range(M):
    axes[i].plot(t, filters[:, i], 'b')
    axes[i].stem(tau[:, i], alpha[:, i], 'r', markerfmt=" ")
```

An example of 48kHz room simulations:

```python
SAMPLING_FREQUENCY = 48000
[filters, alpha, tau] = measurement_tools.load_simulated_rirs(M, K, SAMPLING_FREQUENCY, file_number)

# sort the locations and the weights according to the order of locations
t = np.arange(0, filters.shape[0])/SAMPLING_FREQUENCY
delta = np.zeros((M,1))
fig, axes = plt.subplots(M, 1, figsize=(20,10))
for i in range(M):
    axes[i].plot(t, filters[:, i], 'b')
    axes[i].stem(tau[:, i], alpha[:, i], 'r', markerfmt=" ")
    delta[i] = np.min([abs(tau[:,i] - np.roll(tau[:,i],1)), abs(tau[:,i] - np.roll(tau[:,i],-1))])
```
