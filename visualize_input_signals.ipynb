```python
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
import os
import numpy as np
import scipy.fftpack

import measurement_tools
```

An example of 16kHz speech signals:

```python
T = 1 # signal lenght
F = 16000
t = np.linspace(0.0, T, F*T)
f = np.linspace(0.0, F/2, 1)
fig, ax = plt.subplots(5,2,  constrained_layout=True)
fig.set_size_inches(11,8)
for i in range(5):
    y = measurement_tools.load_input_signal(T, F, i + 1)

    yf = scipy.fftpack.fft(y)
    xf = np.linspace(0.0, 1.0/(2.0*T), len(y)/2)
    N = len(y)
    ax[i,0].plot(t, y)#xf, 2.0/N * np.abs(yf[:N//2]))
    ax[i,0].set_title('Original signal')

    ax[i,1].plot(xf*F, 2.0/N * np.abs(yf[:N//2]))
    ax[i,1].set_title('Zoomed spectrum')
    ax[i,1].set_xlim(0, 5000)
plt.show()
```
